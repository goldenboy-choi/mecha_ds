```python
# 전처리 2가지

# 1번 표준화 (데이터 집합의 분산과 표준편차를 1로 만들어주는 작업)
# 2번 Min Max (데이터의 최대값을 1로 최소값을 0으로 만들어주는 작업)

# 1. 표준화
from sklearn.preprocessing import StandardScaler # 표준화 라이브러리 호출

# 표준화(StandardScaler) 작업을 수행하는 SC00이라는 객체를 생성 및 정의
SC00 = StandardScaler()
# SC00은 임의 네이밍 SS, Sscaler등 본인이 정의

# 표준화 할 데이터를 넣고 피팅 학습
new_data = SC00.fit_transform(df_train2)

# 데이터가 제대로 생성되었는지 확인 (array로 리턴됨)


new_data

```

```python
# dataframe을 합치는 문법
# 대표적으로 merge 와 concat이 있으며, 금일은 concat 소개


# 1번 dataframe
final_df

# 2번 dataframe : 기존 csv를 호출 한 df_train에서 2개 칼럼만 취출하여 dataframe 생성
temp = df_train[['PTC_ID','EYE_CLOSED']]

# 위의 1번과 2번을 merge함 (디폴트는 행으로 합쳐짐.axis = 0 이라는 의미 // 여기서는 열로 합쳐야 하니 axis = 1 로 처리)
a_final = pd.concat([final_df,temp],axis=1)
a_final

# 여기서부터 문제가 시작 임
```

```python
# 위의 결과에서 AF3-AF4행의 평균을 구해서 별도 공간에 담아둬라
# 배열 / 튜플 / 딕셔너리 어떤걸 사용해도 무방
# 이번 시간은 딕셔너리를 사용하여 칼럼들의 평균을 담고자 함

# 딕셔너리 정의
dic_mean = {}

# for문이 돌면서 각 칼럼들의 평균을 계산하여 dic_mean 보관
for col in cols : 
    
    #print(df_train[col].mean()) # 출력하는 부분
    dic_mean[col] = df_train[col].mean()
```

```python
# 위에서 계산된 값을 조회 : 딕셔너리
dic_mean
```

```python
# dataframe으로 변환
dic_mean.keys()
dic_mean.values()

df_from_dic = pd.DataFrame({'Column_Name(칼럼명)' : dic_mean.keys(), 'Mean(평균)' : dic_mean.values() })
df_from_dic
```

```python
# 위에서 최대값
MAX = df_from_dic.max()['Mean(평균)']
MAX
```

```python
# 위에서 최소값
MIN = df_from_dic.min()['Mean(평균)']
MIN
```

```python
# 최대값 - 최소값
RESULT = MAX - MIN
RESULT
```
