```python

# pandas라는 API (라이브러리)를 호출하여 pd라는 별칭으로 사용 한다.
import pandas as pd  

# lv-2308(실제 기출문제 파일) 이라는 csv 파일을 가져와서 df라는 dataframe에 담아둔다

# 개인 피씨의 파일을 불러오는 경우
df  = pd.read_csv('D:/DataScience/Data Science Level2 Practical Problem 2308/lv2-2308.csv') 

# 파이썬 쥬피터 사용하는 경우
#df  = pd.read_csv('./elec/lv2-2308.csv') # 

# 위에서 만든 df라는 dataframe을 보여줘라
display(df)
```

```python
# dataframe에서 기초 통계량

# 1. 칼럼들의 평균을 구해라
df.mean() 
```

```python
# 2. 칼럼들의 최대값을 구해라
df.max()
```

```python
# 칼럼별로 최소값을 구해라
df['AF3'].mean() 
df['F7'].mean() 
df['F3'].mean() 
df['FC5'].mean() 
df['T7'].mean() 
df['P7'].mean() 
df['O1'].mean() 
df['O2'].mean() 
df['P8'].mean() 
df['T8'].mean() 
df['FC6'].mean() 
df['F4'].mean() 
df['F8'].mean() 
df['AF4'].mean() 

# 위와 같은 방법으로는, 손이 많이가고 번거로움
```

```python
# for문을 통해서 계산

# 1. 딕셔너리에 담기

# 튜플 선언 (dic_mean 이라는 딕셔너리를 하나 만들겠다)
dic_mean = {}


# 위에 생성한 df라는 dataframe의 칼럼값의 배열에서 df_cols에 넣는다. → 리스트로 변환한다. 
# 총 16개의 칼럼 data가 df_cols라는 배열 안에 들어간다..
df_cols = list(df.columns)

# df_cols 개수인 16개 만큼 for문을 돌려서, 해당 칼럼값에 해당 하는 평균을 구한뒤 tuple_mean에 결과를 담는다
for col in df_cols :
    dic_mean[col] = df[col].mean()
    
# 계산된 tuple_mean을 출력한다.
# 칼럼별 평균값이 딕셔너리에 들어가 있다. 
dic_mean
```

```python
# 우리가 사용하기 익숙한 dataframe으로 변환
# 1. 라이브러리 사용 (오늘 교육에는 언급 못했습니다.) 
df_from_dic1 = pd.DataFrame.from_dict(dic_mean, orient='index')
df_from_dic1
```

```python
# 여태 배운 내용을 복기 삼아 직접 dataframe을 생성

col_name = list(dic_mean.keys()) # 칼렴명 16개를 col_name이라는 배열에 담아둔다
mean_value = list(dic_mean.values()) # 평균값 16개를 mean_value이라는 배열에 담아둔다

df_from_dic1 = pd.DataFrame({'칼럼명' : col_name,'평균' : mean_value})
df_from_dic1
```

```python
# 참고 : 평균값 이외에 최대값, 최소값, 중앙값, 범위 등 원하는 칼럼을 자유롭게 생성

# 최대값, 최대값, 개수 데이터를 보관할 배열 정의
add_max = []
add_min = []
add_median = []
add_cnt = []

for col in df_cols :
    add_max.append(df[col].max()) # add_max 배열안에 칼럼별 최대값을 보관
    add_min.append(df[col].min()) # add_min 배열안에 칼럼별 최소값을 보관
    add_median.append(df[col].median()) # add_min 배열안에 칼럼별 최소값을 보관
    add_cnt.append(df[col].count()) # add_cnt 배열안에 칼럼별 갯수를 보관
    
# 기존 df_from_dic1에 새로운 칼럼을 추가해서 위에서 산출한 통계량을 넣는다.    
df_from_dic1['최대']  = add_max
df_from_dic1['최소']  = add_min
df_from_dic1['중앙값']  = add_median
df_from_dic1['개수']  = add_cnt

df_from_dic1
```

```python
# 데이터 프레임 소팅 하는 방법

# 평균 값을 기준으로 오름 차순
df_from_dic1.sort_values('평균',ascending=True)
```

```python
# 평균 값을 기준으로 내림 차순
df_from_dic1.sort_values('평균',ascending=False)
```

```python
# 평균 칼럼의 최대값과 최소값의 차이를 구해라. (실제 시험 문제 형식)
answer = df_from_dic1['평균'].max() - df_from_dic1['평균'].min()
answer
```

```python
# 소수점 2자리 까지 보여주라
round(answer,2)
```
