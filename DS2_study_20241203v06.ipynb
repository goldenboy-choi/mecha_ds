```python
# 판다스를 통하여 데이터 가져오기
import pandas as pd 

# 데이터를 가져와서 데이터프레임에 담기 (개인피씨)
df_2308 = pd.read_csv('D:/DataScience/Data Science Level2 Practical Problem 2308/lv2-2308.csv')

# 데이터를 가져와서 데이터프레임에 담기 (쥬피터노트북)
#df_2308  = pd.read_csv('./elec/lv2-2308.csv') 

df_2308
```

```python
# 조회방법
# location

# df_2308에서 EYE_CLOSED가 0인 데이터만 조회해서, df_eye0 에 담아라
df_eye0 = df_2308.loc[df_2308['EYE_CLOSED'] == 0]

# df_2308에서 EYE_CLOSED가 0인 데이터만 조회해서,  df_eye1 에 담아라
df_eye1 = df_2308.loc[df_2308['EYE_CLOSED'] == 1]
```

```python
# 데이터 프레임에서 칼럼명들을 배열로 보관하는 방법

# 전체 칼럼 → 인덱스로 반환
df_2308.columns
```

```python
# 전체 칼럼 → 인덱스로 반환 → 리스트(배열)로 변환
df_2308.columns.tolist()
```

```python
# 특정 칼럼만 조회해서 배열에 담으려고 함 (AF3부터 AF4까지)

column_list = df_2308.loc[:,'AF3':'AF4'].columns.tolist()
column_list
```

```python
# 하나하나 텍스트를 입력하는 방법도 있지만, 과다시간 소요 / 휴먼 에러 가능성
colnumn_list2 = ['AF3','F7','F3','FC5','T7','P7','O1','O2','P8','T8','FC6','F4','F8','AF4']
```

```python
# EYE_CLOSED 가 0인 데이터를 대상으로 14개 칼럼값의 평균을 계산하고자 함

# 14개 칼럼의 평균을 보관할 배열
list_eye0_mean = []

for col in column_list :

    temp_mean = df_eye0[col].mean()

    # 화면에 조회
    print(col,temp_mean)

    # 평균만 별도 배열에 보관
    list_eye0_mean.append(temp_mean)
```

```python
# EYE_CLOSED 가 1인 데이터를 대상으로 14개 칼럼값의 평균을 계산하고자 함

# 14개 칼럼의 평균을 보관할 배열
list_eye1_mean = []

for col in column_list :

    temp_mean = df_eye1[col].mean()

    # 화면에 조회
    print(col,temp_mean)

    # 평균만 별도 배열에 보관
    list_eye1_mean.append(temp_mean)
```

```python
# 3개의 배열이 만들어 졌다
# 1. 칼럼명 14개  → column_list
# 2. EYE_CLOSED 가 0인 경우의 칼럼들의 평균 → list_eye0_mean
# 3. EYE_CLOSED 가 1인 경우의 칼럼들의 평균 → list_eye1_mean

# 배열 3개로, 데이터프레임을 생성 한다.. 직관적..

# 3. EYE_CLOSED 가 1인 경우의 칼럼들의 평균 → list_eye1_mean
last_df = pd.DataFrame({'COLUMN': column_list, 'EYE0_MEAN' : list_eye0_mean, 'EYE1_MEAN' : list_eye1_mean})
last_df
```
